```python
import pandas as pd

def sort_and_concatenate_ids(row: pd.Series) -> str:
    """
    Sorts flight_id1 and flight_id2 alphanumerically and concatenates them.

    Args:
        row (pd.Series): A row of a DataFrame with 'flight_id1' and 'flight_id2'.

    Returns:
        str: Concatenated string of sorted flight IDs.
    """
    sorted_ids = sorted([str(row['flight_id1']), str(row['flight_id2'])])
    return "_".join(sorted_ids)

# Apply the function row-wise

pairs_df['sorted_concat'] = pairs_df.apply(sort_and_concatenate_ids, axis=1)
ce_h3_half_df['sorted_concat'] = ce_h3_half_df.apply(sort_and_concatenate_ids, axis=1)
f =  [not x for x in pairs_df.sorted_concat.isin(ce_h3_half_df.sorted_concat).to_list()]
pairs_df[f]
```
